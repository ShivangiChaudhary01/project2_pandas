```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# read data from csv file into a dataframe

```python
data_1 = pd.read_csv("dataset_1.csv")
data_1
```

```python
#understanding the data for data cleaning by having data summary
# using shape,display info and describe for the summary
```

```python
data_1.shape
```

```python
print("Displaying the info of dataset")
print(data_1.info())
```

```python
data_1.describe()
```

```python
data_1.isnull().sum()
```

```python
#dropping nul values if preesent

data_1.dropna(inplace =True)
data_1
```

```python
#here i am checking,whether there is any duplicate rows or columns in the dataset
#As there are no such duplicates,we will not perform the code to remove duplicates

duplicate_rows = data_1.duplicated()
print("Number of duplicate rows:", duplicate_rows.sum())
duplicate_columns = data_1.duplicated()
print("Number of duplicate columns:", duplicate_columns.sum())
```

```python
#checking unique values for each column
# Count of unique values per column

unique_counts = data_1.nunique()

print(unique_counts)
```

```python
#changing the dtype of dteday as it can not be object. so converting it into datetime.
#other columns dtypes are correct so we will not change them.

data_1['dteday'] = pd.to_datetime(data_1['dteday'], format='%d-%m-%Y')
print(data_1)
```

```python
data_1['holiday'] = data_1['holiday'].map({'FALSE': False, 'TRUE': True})
data_1['holiday'] = data_1['holiday'].astype(bool)
```

```python
print(data_1['holiday'].dtype)
```

```python
#Dropping the 'yr','mnth'column, because it is of no use as dteday column is giving the information .

finaldata_1 = data_1.drop(columns=['yr','mnth'])
print(finaldata_1)
```

# making some EDAs for better understanding for dataset.

# What 1st chart shows:
Each bar represents a weekday (0=Sunday, 1=Monday, ..., 6=Saturday).

The height of the bar shows how many times a holiday occurs on that weekday (in your dataset, each row is an hour, so this is the number of holiday hours per weekday).

# The second graph gives a clear, visual comparison between holidays and non-holidays for each weekday

# the third graph shows the average temperature for each season.Also, have found the stats in beginning of code under summary markdown.

```python
# Count number of holidays per weekday
holiday_counts = finaldata_1[finaldata_1['holiday']].groupby('weekday').size()

# Plot
holiday_counts.plot(kind='bar')
plt.title('Number of Holidays per Weekday')
plt.xlabel('Weekday (0=Sunday, 6=Saturday)')
plt.ylabel('Number of Holiday Hours')
plt.show()
```

```python
# Create a crosstab of holidays by weekday
crosstab = pd.crosstab(finaldata_1['weekday'], finaldata_1['holiday'])

# Plot
crosstab.plot(kind='bar', stacked=True)
plt.title('Holiday vs Non-Holiday Hours per Weekday')
plt.xlabel('Weekday (0=Sunday, 6=Saturday)')
plt.ylabel('Number of Hours')
plt.legend(['Non-Holiday', 'Holiday'])
plt.show()
```

```python
#Dropping the 'Unnamed: 0' column, because it is of no use and has same content as instant column.

finaldata_2 = data_2.drop(columns=['Unnamed: 0'])
print(finaldata_2)
```

# making some EDAs for better understanding for dataset.

```python
# Line graph of total count over time
#this graph shows Trends over time (Line graph): Reveals growth, seasonality, or sudden changes in bike usage.


plt.figure(figsize=(10,5))
plt.plot(finaldata_2['instant'], finaldata_2['cnt'], marker='o', linestyle='-')
plt.title('Total Count Over Time')
plt.xlabel('Instant')
plt.ylabel('Total Count (cnt)')
plt.grid(True)
plt.show()
```

```python
# Sum registered and casual users
#This chart shows User segmentation (Bar graph): Highlights the proportion of regular vs. casual riders, useful for business decisions.


totals = [finaldata_2['registered'].sum(), finaldata_2['casual'].sum()]
labels = ['Registered', 'Casual']

plt.bar(labels, totals, color=['orange', 'green'])
plt.title('Total Registered vs. Casual Users')
plt.ylabel('Total Users')
plt.show()
```

```python
#This line (or scatter) graph shows how the total count (cnt) changes with humidity (hum).
#it will help in Behavior patterns: Identifies optimal conditions for bike use and potential barriers (like extreme weather).


plt.plot(finaldata_2['hum'], finaldata_2['cnt'], 'bo', alpha=0.5)
plt.title('Humidity vs. Total Count')
plt.xlabel('Humidity')
plt.ylabel('Total Count (cnt)')
plt.show()
```

# we will merge our data_1 and data_2 and make new dataset_A

```python
dataset_A= pd.merge(finaldata_1,finaldata_2,on="instant",how="inner")
dataset_A
```

```python
# Suppose your new dataset is stored in a variable called `new_dataset`
dataset_A.to_csv('dataset_A.csv', index=False)
```
