# OUTLIER DETECTION + SKEWNESS & KURTOSIS

```python
from scipy.stats import skew, kurtosis

# 📌 Step 1: Skewness and Kurtosis
for col in ['temp', 'hum', 'windspeed', 'cnt']:
    print(f"{col} - Skewness: {skew(master_data[col])}, Kurtosis: {kurtosis(master_data[col])}")
```

# Handle Outliers using IQR Method

```python


# Copy the original before trimming
master_data_iqr = master_data.copy()

def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[column] >= lower) & (data[column] <= upper)]

# Apply IQR filter to each numerical column
for col in ['temp', 'hum', 'windspeed', 'cnt']:
    master_data_iqr = remove_outliers_iqr(master_data_iqr, col)

print("Original shape:", master_data.shape)
print("After IQR cleaning:", master_data_iqr.shape)
```

```python
#Boxplot after Outlier Removal

for col in ['temp', 'hum', 'windspeed', 'cnt']:
    plt.figure(figsize=(6, 4))
    sns.boxplot(y=master_data_iqr[col], color='limegreen')
    plt.title(f"Boxplot of {col} (After IQR Cleaning)")
    plt.show()
```
